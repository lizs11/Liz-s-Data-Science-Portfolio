## Outliers

Removing outliers from the data is discarding values that lie a long way from the median or mean value. For this dataset, any rows where the income is >50K but the age is below 22 or above 72 - will be removed, basically saying that these values are reasonably unlikely to be real.   


```python
print( len( adult[(adult.income == '>50K')] ))
print( len( adult[(adult.age > 72)] ))
print( len( adult[(adult.age < 22)] ))
print( len( adult[(adult.income == '>50K')  & ((adult.age > 72) | (adult.age < 22))] ))
```

Based on these expressions, we can drop rows outside of the target range using the `drop` method.  Note the use of `adult.index`.  The argument to drop is a series giving the indexes of the rows we want to remove from the dataframe.   So we filter the index of the dataframe using the boolean expression to get this.

```python
# remove outliers - people with high income who are too old or too young
adult_clean = adult.drop( adult.index[(adult.income == '>50K') & 
                                      ((adult['age'] > 72) | 
                                       (adult['age'] < 22))]
                        )


# confirm that we have restricted the range of ages for high earners, should see 
adult_clean[adult_clean.income=='>50K'].age.describe()
```

Finally, use the **Pearson’s median skewness coefficient** to confirm the difference in the shape of the distribution of male and female age distributions. 

```python

def skewness(x):
    res = 0
    m = x.mean()
    s = x.std()
    for i in x:
        res += (i-m) * (i-m) * (i-m)
    res /= (len(x) * s * s * s)
    return res

print ("Skewness of the male population = ", skewness(males.age))
print ("Skewness of the female population = ", skewness(females.age))

print ("Skewness of the cleaned male population = ", skewness(males_clean.age))
print ("Skewness of the cleaned female population = ", skewness(females_clean.age))

```

```python
def pearson(x):
    return 3*(x.mean() - x.median())*x.std()

print ("Pearson's coefficient of the male population = ", pearson(males.age))
print ("Pearson's coefficient of the female population = ", pearson(females.age))

print ("Pearson's coefficient of the cleaned male population = ", pearson(males_clean.age))
print ("Pearson's coefficient of the cleaned female population = ", pearson(females_clean.age))
```
