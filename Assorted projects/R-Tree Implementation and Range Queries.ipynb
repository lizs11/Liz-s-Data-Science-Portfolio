```python
class Node(object): # node class
    def __init__(self):
        self.id = 0
        # for internal nodes
        self.child_nodes = []
        # for leaf nodes
        self.data_points = []
        self.parent = None
        self.MBR = {
            'x1': -1,
            'y1': -1,
            'x2': -1,
            'y2': -1,
        }
    def perimeter(self):
        # only calculate the half perimeter here
        return (self.MBR['x2'] - self.MBR['x1']) + (self.MBR['y2'] - self.MBR['y1'])

    def is_overflow(self):
        if self.is_leaf(): # check for leaf node
            if self.data_points.__len__() > B: # Checking overflows of data points, B is the upper bound.
                return True
            else:
                return False
        else:
            if self.child_nodes.__len__() > B: # Checking overflows of child nodes, B is the upper bound.
                return True
            else:
                return False

    def is_root(self):
        if self.parent is None: # check if there is no parent node
            return True
        else:
            return False

    def is_leaf(self):
        if self.child_nodes.__len__() == 0: # check if there are no child nodes
            return True
        else:
            return False
```

```python
class RTree(object): # R tree class
    def __init__(self):
        self.root = Node() # Create a root

    def query(self, node, query): # run to answer the query
        num = 0
        if node.is_leaf(): # check if a data point is included in a leaf node
            for point in node.data_points:
                if self.is_covered(point, query): # if the point is covered by the query
                    num = num + 1
            return num
        else:
            for child in node.child_nodes: # If it is an MBR, check all the child nodes to see if there is an interaction
                if self.is_intersect(child, query): # If there is an interaction, continue to check the child nodes in the next layer through to the leaf nodes
                    num = num + self.query(child, query)
            return num

    def is_covered(self, point, query):
        x1, x2, y1, y2 = query['x1'], query['x2'], query['y1'], query['y2']
        if x1 <= point['x'] <= x2 and y1 <= point['y'] <= y2: # if the point is included in the query 
            return True
        else:
            return False    

    def is_intersect(self, node, query): 
        # if two mbrs are intersected, then:
        # |centre1_x - center2_x| <= length1 / 2 + length2 / 2 and:
        # |centre1_y - center2_y| <= width1 / 2 + width2 / 2
        centre1_x = (node.MBR['x2'] + node.MBR['x1']) / 2 # find the centre points
        centre1_y = (node.MBR['y2'] + node.MBR['y1']) / 2
        length1 = node.MBR['x2'] - node.MBR['x1'] # find the lengths of the MBRs
        width1 = node.MBR['y2'] - node.MBR['y1']
        centre2_x = (query['x2'] + query['x1']) / 2
        centre2_y = (query['y2'] + query['y1']) / 2
        length2 = query['x2'] - query['x1']
        width2 = query['y2'] - query['y1'] 
        if abs(centre1_x - centre2_x) <= length1 / 2 + length2 / 2 and\
                abs(centre1_y - centre2_y) <= width1 / 2 + width2 / 2:  # check the absolute value
            return True
        else:
            return False                    


    def insert(self, u, p): # insert p(data point) to u (MBR)
        if u.is_leaf(): # check for leaf node
            self.add_data_point(u, p) # add the data point and update the corresponding MBR
            if u.is_overflow(): # check if the node has overflowed
                self.handle_overflow(u) # handle overflow for leaf nodes
        else:
            v = self.choose_subtree(u, p) # choose a subtree to insert the data point to minimise the perimeter sum
            self.insert(v, p) # continue to check the next layer recursively
            self.update_mbr(v) # update the MBR for inserting the data point

    def choose_subtree(self, u, p): 
        if u.is_leaf(): # find the leaf and insert the data point
            return u
        else:
            min_increase = sys.maxsize # set an initial value
            best_child = None
            for child in u.child_nodes: # check each child to find the best node to insert the point 
                if min_increase > self.peri_increase(child, p): # check if minimum increase is greater than the perimeter increase
                    min_increase = self.peri_increase(child, p)
                    best_child = child
            return best_child

    def peri_increase(self, node, p): # calculate the increase of the perimeter after inserting the new data point
        # new perimeter - original perimeter = increase of perimeter
        origin_mbr = node.MBR
        x1, x2, y1, y2 = origin_mbr['x1'], origin_mbr['x2'], origin_mbr['y1'], origin_mbr['y2']
        increase = (max([x1, x2, p['x']]) - min([x1, x2, p['x']]) +
                    max([y1, y2, p['y']]) - min([y1, y2, p['y']])) - node.perimeter()
        return increase


    def handle_overflow(self, u):
        u1, u2 = self.split(u) # u1 and u2 are the two splits returned by the function "split"
        # if u is root, create a new root with u1 and u2 as its children
        if u.is_root(): # check if u is a root
            new_root = Node()
            self.add_child(new_root, u1)
            self.add_child(new_root, u2)
            self.root = new_root
            self.update_mbr(new_root)
        # if u is not root, delete u, and set u1 and u2 as u's parent's new children
        else:
            w = u.parent
            # copy the information of u1 into u
            w.child_nodes.remove(u)
            self.add_child(w, u1) # link the two splits and update the corresponding MBR
            self.add_child(w, u2)
            if w.is_overflow(): # check the parent node recursively
                self.handle_overflow(w)
            
    def split(self, u):
        # split u into u1 and u2
        best_u1 = Node()
        best_u2 = Node()
        best_perimeter = sys.maxsize
        # u is a leaf node
        if u.is_leaf():
            m = u.data_points.__len__()
            # create two different kinds of divides
            divides = [sorted(u.data_points, key=lambda data_point: data_point['x']),
                       sorted(u.data_points, key=lambda data_point: data_point['y'])] # sorting the points based on X and Y dimensions
            for divide in divides: # for all divides
                for i in range(math.ceil(0.4 * B), m - math.ceil(0.4 * B) + 1): # check the combinations to find a near-optimal one
                    u1 = Node()
                    u1.data_points = divide[0: i]
                    self.update_mbr(u1)
                    u2 = Node()
                    u2.data_points = divide[i: divide.__len__()]
                    self.update_mbr(u2)
                    if best_perimeter > u1.perimeter() + u2.perimeter(): # if the best perimeter is greater than the sum of the two splits' perimeters
                        best_perimeter = u1.perimeter() + u2.perimeter()
                        best_u1 = u1
                        best_u2 = u2

        # u is an internal node
        else:
            # create four different kinds of divides
            m = u.child_nodes.__len__()
            divides = [sorted(u.child_nodes, key=lambda child_node: child_node.MBR['x1']), # sorting based on MBRs
                       sorted(u.child_nodes, key=lambda child_node: child_node.MBR['x2']),
                       sorted(u.child_nodes, key=lambda child_node: child_node.MBR['y1']),
                       sorted(u.child_nodes, key=lambda child_node: child_node.MBR['y2'])]
            for divide in divides: # for all divides
                for i in range(math.ceil(0.4 * B), m - math.ceil(0.4 * B) + 1): # check the combinations
                    u1 = Node()
                    u1.child_nodes = divide[0: i]
                    self.update_mbr(u1)
                    u2 = Node()
                    u2.child_nodes = divide[i: divide.__len__()]
                    self.update_mbr(u2)
                    if best_perimeter > u1.perimeter() + u2.perimeter(): # if the best perimeter is greater than the sum of the two splits' perimeters
                        best_perimeter = u1.perimeter() + u2.perimeter()
                        best_u1 = u1
                        best_u2 = u2

        for child in best_u1.child_nodes: # for all child nodes in the best splits, set parent node
            child.parent = best_u1
        for child in best_u2.child_nodes:
            child.parent = best_u2

        return best_u1, best_u2


    def add_child(self, node, child):
        node.child_nodes.append(child) # add child nodes to the current parent (node) and update the MBRs. Used in handling overflows
        child.parent = node
        if child.MBR['x1'] < node.MBR['x1']: # if the child MBRs are greater than the node MBRs
            node.MBR['x1'] = child.MBR['x1']
        if child.MBR['x2'] > node.MBR['x2']:
            node.MBR['x2'] = child.MBR['x2']
        if child.MBR['y1'] < node.MBR['y1']:
            node.MBR['y1'] = child.MBR['y1']
        if child.MBR['y2'] > node.MBR['y2']:
            node.MBR['y2'] = child.MBR['y2']
    # return the child whose MBR requires the minimum increase in perimeter to cover p

    def add_data_point(self, node, data_point): # add data points and update MBRs
        node.data_points.append(data_point)
        if data_point['x'] < node.MBR['x1']: # if the point MBRs are greater than the node MBRs
            node.MBR['x1'] = data_point['x']
        if data_point['x'] > node.MBR['x2']:
            node.MBR['x2'] = data_point['x']
        if data_point['y'] < node.MBR['y1']:
            node.MBR['y1'] = data_point['y']
        if data_point['y'] > node.MBR['y2']:
            node.MBR['y2'] = data_point['y']


    def update_mbr(self, node): # update MBRs when forming a new MBR. Used in checking combinations and updating root
        x_list = []
        y_list = []
        if node.is_leaf(): # if the node is a leaf node
            x_list = [point['x'] for point in node.data_points]
            y_list = [point['y'] for point in node.data_points]
        else:
            x_list = [child.MBR['x1'] for child in node.child_nodes] + [child.MBR['x2'] for child in node.child_nodes]
            y_list = [child.MBR['y1'] for child in node.child_nodes] + [child.MBR['y2'] for child in node.child_nodes]
        new_mbr = {
            'x1': min(x_list),
            'x2': max(x_list),
            'y1': min(y_list),
            'y2': max(y_list)
        }
        node.MBR = new_mbr
```
